```python
# library
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = 'http://www.allocine.fr/film/meilleurs/'
```

```python
r = requests.get(url).content
soup = BeautifulSoup(r,'html')
#soup
```

```python
# all the information
movie_box = soup.find_all('li',{'class':'mdl'})
#movie_box
```

#### rank

```python
#rank
rank = soup.select('li.mdl div.label')
lst_rank = [i.text for i in soup.select("li.mdl div.label")]
lst_rank
```

#### title

```python
#title
title = soup.select('div.meta h2')
lst_title = [x.text.strip() for x in soup.select('div.meta h2')]
lst_title
```

#### genre

```python
#genre
genre = soup.select("div.meta span[class*='w==']")
#genre
title = soup.select('div.meta')
lst_genre = [[i.text for i in title[0].find_all('span')],
        [i.text for i in title[1].find_all('span')],
        [i.text for i in title[2].find_all('span')],
        [i.text for i in title[3].find_all('span')],
        [i.text for i in title[4].find_all('span')],
        [i.text for i in title[5].find_all('span')],
        [i.text for i in title[6].find_all('span')],
        [i.text for i in title[7].find_all('span')],
        [i.text for i in title[8].find_all('span')],
        [i.text for i in title[9].find_all('span')]]
lst_genre
```

```python
del lst_genre[1][2]
del lst_genre[9][2]
lst_genre
```

#### director

```python
#director
director = soup.select("div.meta-body-direction")
lst_director = [i.text.replace('\n','').replace('  ','').replace('De','')
                .split(',') for i in soup.select("div.meta-body-direction")]
lst_director
```

#### actor

```python
#actor
#actor = soup.select('div.meta-body-actor a[href*=personne]')
lst_actor = [i.text.replace('\n','').replace('  ','').replace('Avec','')
             .split(',') for i in soup.select("div.meta-body-actor")] 
lst_actor
```

#### press & spectator notes

```python
#note spectators + press
p_s = soup.select("div.rating-item-content span.stareval-note")
#p_s
lst_p_s = [x.text.strip() for x in soup.select("div.rating-item-content span.stareval-note")]
lst_p_s
```

```python
#divide press [0], spectator [1]
split_lst = (list(zip(*[iter(lst_p_s)]*2)))
split_lst
```

```python
lst_press = list(zip(*split_lst))[0]
lst_press
```

```python
lst_spect = list(zip(*split_lst))[1]
lst_spect
```

```python
lst_rank1 = []
lst_title1 = []
lst_genre1 = []
lst_director1 = []
lst_actor1 = []
lst_press1 = []
lst_spectator1 = []

for i in range(1,31):
    r = requests.get(f'http://www.allocine.fr/film/meilleurs/?page={i}')    
    soup = BeautifulSoup(r.content,'lxml') 
    
    lst_rank1.append([x.text for x in soup.select("li.mdl div.label")])
    lst_title1.append([x.text.strip() for x in soup.select('div.meta h2')])
    
    title = soup.select('div.meta')
    lst_genre1.append([[i.text for i in title[0].find_all('span')],
                [i.text for i in title[1].find_all('span')],
                [i.text for i in title[2].find_all('span')],
                [i.text for i in title[3].find_all('span')],
                [i.text for i in title[4].find_all('span')],
                [i.text for i in title[5].find_all('span')],
                [i.text for i in title[6].find_all('span')],
                [i.text for i in title[7].find_all('span')],
                [i.text for i in title[8].find_all('span')],
                [i.text for i in title[9].find_all('span')]])
    #del lst_genre[1][2]
    #del lst_genre[9][2]
    
    lst_director1.append([i.text.replace('\n','').replace('  ','').replace('De','')
                .split(',') for i in soup.select("div.meta-body-direction")])
    lst_actor1.append([i.text.replace('\n','').replace('  ','').replace('Avec','')
             .split(',') for i in soup.select("div.meta-body-actor")])
    
    lst_p_s = [x.text.strip() for x in soup.select("div.rating-item-content span.stareval-note")]
    split_lst = (list(zip(*[iter(lst_p_s)]*2)))
    lst_press1.append(list(zip(*split_lst))[0])
    lst_spectator1.append(list(zip(*split_lst))[1]) 
```
