# Clean Data

- drop "Unnamed: 0'" column : i don't understand what is it 
- check if there are missing values and drop rows or columns if it's necessary
- check the type of the data

```python
# drop column 
df = df.drop('Unnamed: 0',axis=1)
df = df.drop('sup_devices.num',axis=1)
df = df.drop('ipadSc_urls.num',axis=1)
df = df.drop('lang.num',axis=1)
df = df.drop('vpp_lic',axis=1)
```

```python
# check NaN
df.isna().sum()
# No missing values 
```

```python
df.dtypes
```

```python
# what contains currency column ? 
#df.currency.unique()
# There is just one value : USD
# don't need it, drop it
df = df.drop('currency', axis = 1)
```

```python
# price : 
#df.price.unique()
# create binary column : 1 = not free & 0 = free 
df['free_notfree'] = df.price.apply(lambda x : 0 if x == 0 else 1)
# drop price column 
#df = df.drop('price', axis = 1)
```

```python
# drop ver(sion) column 
df = df.drop('ver', axis = 1)
```

```python
# cont_rating column : age for each app 
#df.cont_rating.unique()
# need to drop '+' and change the type to numeric 
# create a fonction to convert string to numeric & drop '+'
def convert_string_to_numeric(s):
    if (type(s)==str) and (s[-1] == '+'):
            result = int(s[:-1])
            return result
# apply the fonction :
df['cont_rating'] = df.cont_rating.apply(lambda s : convert_string_to_numeric(s))
```

```python
# change the size of size's column 
df['size_MB'] = df['size_bytes'] / (1024 * 1024.0)
df = df.drop('size_bytes', axis = 1)
```

def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out

# Logistic Regression

to predict the success of an app depending the price, the size, the genre : to create a column 0 & 1
0 = 0,1,1.5,2,2.5
1 = 3,3.5,4,4.5,5

```python
dflog = df.copy()
```

```python
dflog = dflog.drop('id',axis=1).drop('track_name',axis=1).drop('price',axis=1).drop(
    'rating_count_tot',axis=1).drop('rating_count_ver',axis=1).drop(
    'user_rating_ver',axis=1).drop('cont_rating',axis=1)
```

```python
# dummies genres' column 
dflog = pd.get_dummies(dflog, columns=['prime_genre'])
```

```python
dflog.head()
```

```python
# create a column with 1 (3-5) for success & 0 (0-2.5) for no success
# what's in this column ? 
dflog.user_rating.unique()
#array([4. , 4.5, 3.5, 5. , 3. , 2. , 2.5, 0. , 1.5, 1. ])
dflog['success_ornot'] = np.where(dflog['user_rating']<2.5, 1,0)
# and now : 
dflog['success_ornot'].value_counts()
#0    5721
#1    1042
#Name: success_ornot, dtype: int64
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
sns.countplot(x='success_ornot',data=dflog,palette="Set2")
plt.title('Success or not',fontsize=16)
plt.xlabel('Not = 0 & Success = 1', fontsize=16)
plt.show()
```

```python
#dflog = dflog.drop('size_MB',axis=1)
```

```python
dflog = dflog.drop('user_rating',axis=1) #  R-squared: 0.043
```

```python
X = dflog.iloc[:,:17] # Features without target : user rating, size, free, book
y = dflog.success_ornot # Target variable : success or not
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
model = linear_model.LogisticRegression(class_weight='balanced') #change the weight:1value=2rows
result = model.fit(X_train,y_train) 
y_pred=result.predict(X_test)
result.score(X_test,y_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
import statsmodels.api as sm

logit_model=sm.Logit(y_test,X_test)
result=logit_model.fit()
print(result.summary2())
```

## Prediction

```python
# Accuracy score
from sklearn.metrics import accuracy_score

y_pred_test = model.predict(X_test)
acc_test = accuracy_score(y_test, y_pred_test)
print('The accuracy score : ', acc_test)
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix

confusion_matrix = confusion_matrix(y_test, y_pred_test)
print(confusion_matrix)
# TP : 1129 & TN : 220 
# FP : 0 & FN : 4
# The result is telling us that we have 1129+6(1135) correct predictions and 
# 7+212(219) incorrect predictions.
```

```python
# precision + recall + f1-score + support
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_test))

# interpretation ???
```

```python
# Roc curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

logit_roc_auc = roc_auc_score(y_test, model.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```
