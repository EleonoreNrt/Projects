#### create a fonction to group time

```python
import re 

def timeto(t):
    if (re.search('[0][5-9]:',t)) or (re.search('[1][0-1]:',t)):
        t = 'morning'
    elif re.search('[1][2-7]:',t):
        t = 'afternoon'
    elif (re.search('[1][8-9]:',t)) or (re.search('[2][0-1]:',t)):
        t = 'evening'
    elif (re.search('[2][2-4]:',t)) or (re.search('[0][1-4]:',t)):
        t = 'night'
    return t
```

```python
shark['time'] = shark['time'].apply(lambda x : timeto(x) if pd.isnull(x) == False else np.nan) 
```

```python
shark = shark[shark['time'].isin(['morning','afternoon','evening','night'])]
```

#### species

```python
shark['species'] = shark['species'].str.lower().str.strip()
```

```python
shark.species.value_counts() #1433
```

```python
def spec(p):
    if re.search('white',p):
        p = 'white shark'
    elif re.search('tiger',p):
        p = 'tiger shark'
    elif re.search('bull',p):
        p = 'bull shark'
    elif re.search('wobbegong',p):
        p = 'wobbegong shark'
    elif re.search('mako',p):
        p = 'mako shark'
    elif re.search('raggedtooth',p):
        p = 'raggedtooth shark'
    elif re.search('blue',p):
        p = 'blue shark'
    elif re.search('spinner',p):
        p = 'spinner shark'
    elif re.search('whaler',p):
        p = 'whaler shark'
    elif re.search('blacktip',p):
        p = 'blacktip reef shark'
    else :
        p = 'unidentified shark species '
    return p
shark['species'] = shark['species'].apply(lambda x : spec(x) if pd.isnull(x) == False else np.nan) 
```

```python
shark.species.value_counts() #1433
```
